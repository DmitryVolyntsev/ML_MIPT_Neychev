## Часть 1. Работа с моделями и ансамблями моделей в задачи классификации.

### 1. Чтение данных.
Воспользуемся [датасетом](https://archive.ics.uci.edu/ml/datasets/Statlog+%28Vehicle+Silhouettes%29), в котором описываются различные автомобили. Будем решать задачу многоклассовой ($k=4$) классификации.
Для удобства, датасет уже преобразован в удобный формат.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

dataset = pd.read_csv('car_data.csv', delimiter=',', header=None).values
data = dataset[:, :-1].astype(int)
target = dataset[:, -1]

print(data.shape, target.shape)

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.35)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

Для первичного анализа может быть полезна библиотека `pandas`. Преобразуем `train` выборку в `pd.DataFrame`.

```python
X_train_pd = pd.DataFrame(X_train)

# First 15 rows of our dataset.
X_train_pd.head(15)
```

### 2. Работа с данными, построение модели, анализ ошибки.
Выполните следующие манипуляции с данными:

#### 2.0. Предобработка данных.
* Произведите необходимые (по вашему мнению) манипуляции с данными и объясните их.

```python
### YOUR CODE HERE
from sklearn.preprocessing import StandardScaler
```

```python
#отнормируем данные, т.к. значения из большого диапазона и сильно отличаются
skaler = StandardScaler()
X_train_t = skaler.fit_transform(X_train)
X_test_t = skaler.transform(X_test)
```

#### 2.1. Базовая логистическая регрессия.
* Подберите оптимальные параметры логистической регресии с помощью кросс-валидации на train-датасете (е переусердствуйте с подбором, в данной работе не стоит задача найти самую оптимальную модель. Небольшого grid/random search'а хватит).

* Постройте график ROC-кривой для данного классификатора (`sklearn.metrics.roc_curve`), оцените точность классификации и f1-score. 

```python
### YOUR CODE HERE
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from scipy import interp
```

```python
lr_clf = LogisticRegression(solver='saga', multi_class='multinomial', max_iter=10000, n_jobs=-1)
```

```python
pg = {'C': np.linspace(0.01, 1, 10), 'penalty': ['l1', 'l2']}
grid_search = GridSearchCV(estimator=lr_clf, n_jobs=-1, param_grid=pg, cv=5, scoring='accuracy')
grid_search.fit(X_train_t, y_train)

grid_search.best_params_
grid_search.best_estimator_
```

```python
predicted_labels = grid_search.predict(X_test_t)

acc = accuracy_score(y_test, predicted_labels)
f1 = f1_score(y_test, predicted_labels, average='weighted')
print('acc = ', acc)
print('f1 = ', f1)
```

### ROC-кривая

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
class_names = grid_search.best_estimator_.classes_
y_test_b = label_binarize(y_test, class_names)
#вместо predict_labels, полученного методом predict, следует использовать predict_proba
y_predict = grid_search.predict_proba(X_test_t)

fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(len(class_names)):
    fpr[i], tpr[i], _ = roc_curve(y_test_b[:, i], y_predict[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
#для микро
fpr['micro'], tpr['micro'], _ = roc_curve(y_test_b.ravel(), y_predict.ravel())
roc_auc['micro'] = auc(fpr['micro'], tpr['micro'])

#макро
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(len(class_names))]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(len(class_names)):
    mean_tpr += interp(all_fpr, fpr[i], tpr[i]) 
mean_tpr /= len(class_names)
fpr['macro'] = all_fpr
tpr['macro'] = mean_tpr
roc_auc['macro'] = auc(fpr['macro'], tpr['macro'])
```

```python
#графики
plt.figure(figsize=(6,6))
plt.plot(fpr['micro'], tpr['micro'], label='micro, auc = %.3f'%(roc_auc['micro']), linestyle='--')
plt.plot(fpr['macro'], tpr['macro'], label='macro, auc = %.3f'%(roc_auc['macro']), linestyle='--')
for i in range(len(class_names)):
    plt.plot(fpr[i], tpr[i], label='class = %s, auc = %.3f'%(class_names[i], auc(fpr[i], tpr[i])))

plt.title('ROC-Кривая')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

#### 2.6. Bagging.
Перейдем к ансамблям. 

Для построения ансамбля воспользуемся Bagging'ом с числом алгоритмов от 2 до 50. В качестве первого семейства базовых алгоримтов будем использовать линейные модели (т.е. логистическую регрессию), в качестве второго - решающие деревья. 

*Пояснение: Будем строить ансамбль только из моделей из одного семейства, т.е. логистическая регрессия не смешивается с решающими деревьями.*

Для этого можно сгенерировать с помощью метода bootstrap 50 подвыборок из `train` выборки (того же размера, что и исходная), обучить логистическую регрессию и решающее дерево с оптимальными параметрами из предыдущего пункта на каждой из подвыборок и затем усреднить предсказания k моделей.

*Hint: в sklearn может найтись подходящая функция, которая облегчит вам реализацию данного пункта.*

* Постройте график качества классификации и f1-score на `train` и `test` датасетах в зависимости от числа алгоритмов, вошедших в ансамбль.

* Проанализируйте график. Какое количество моделей стоит использовать? Как вы думаете, являются ли параметры решающего дерева, подобранные в предыдущем пункте оптимальными в данном случае?

```python
# YOUR CODE HERE
from sklearn.ensemble import BaggingClassifier
```

LogReg

```python
acc_train = []
acc_test = []
f1_train = []
f1_test = []
for i in range(2, 51):
    lr_clf_again = LogisticRegression(solver='saga', multi_class='multinomial', max_iter=10000, C=1.0, n_jobs=-1)
    bag_log_reg = BaggingClassifier(base_estimator=lr_clf_again, n_estimators=i, bootstrap=True, n_jobs=-1)
    bag_log_reg.fit(X_train_t, y_train)
    class_names = bag_log_reg.classes_
    y_train_b = label_binarize(y_train, class_names)
    y_test_b = label_binarize(y_test, class_names)
    y_predict_train_b = label_binarize(bag_log_reg.predict(X_train_t), class_names)
    y_predict_test_b = label_binarize(bag_log_reg.predict(X_test_t), class_names)
    acc_train.append(accuracy_score(y_train_b, y_predict_train_b))
    f1_train.append(f1_score(y_train_b, y_predict_train_b, average='weighted'))
    acc_test.append(accuracy_score(y_test_b, y_predict_test_b))
    f1_test.append(f1_score(y_test_b, y_predict_test_b, average='weighted'))
```

```python
plt.figure(figsize=(7, 7))
plt.plot(np.arange(2, 51), acc_train, label='accuracy train')
plt.plot(np.arange(2, 51), f1_train, label='f1_score train')
plt.plot(np.arange(2, 51), acc_test, label='accuracy test')
plt.plot(np.arange(2, 51), f1_test, label='f1_score test')
plt.xlabel('n_estimators')
plt.ylabel('score')
plt.legend()
plt.grid()
plt.show()
```

best n_estimators = 26

TreeBagging

```python
acc_train = []
acc_test = []
f1_train = []
f1_test = []
for i in range(2, 51):
    classifier = DecisionTreeClassifier(max_depth=50.5)
    tree_bag = BaggingClassifier(base_estimator=classifier, n_estimators=i, bootstrap=True, n_jobs=-1)
    tree_bag.fit(X_train_t, y_train)
    class_names = tree_bag.classes_
    y_train_b = label_binarize(y_train, class_names)
    y_test_b = label_binarize(y_test, class_names)
    y_predict_train_b = label_binarize(tree_bag.predict(X_train_t), class_names)
    y_predict_test_b = label_binarize(tree_bag.predict(X_test_t), class_names)
    acc_train.append(accuracy_score(y_train_b, y_predict_train_b))
    f1_train.append(f1_score(y_train_b, y_predict_train_b, average='weighted'))
    acc_test.append(accuracy_score(y_test_b, y_predict_test_b))
    f1_test.append(f1_score(y_test_b, y_predict_test_b, average='weighted'))
```

```python
plt.figure(figsize=(7, 7))
plt.plot(np.arange(2, 51), acc_train, label='accuracy train')
plt.plot(np.arange(2, 51), f1_train, label='f1_score train')
plt.plot(np.arange(2, 51), acc_test, label='accuracy test')
plt.plot(np.arange(2, 51), f1_test, label='f1_score test')
plt.xlabel('n_estimators')
plt.ylabel('score')
plt.legend()
plt.grid()
plt.show()
```

best n_estimators = 41

```python
f1_train = []
acc_train = []
f1_test = []
acc_test = []

a = LogisticRegression(solver='saga', multi_class='multinomial', max_iter=10000, C=1.0, n_jobs=-1)
b = DecisionTreeClassifier(max_depth=50.5)

log_reg = LogisticRegression(solver='saga', multi_class='multinomial', max_iter=10000, C=1.0, n_jobs=-1)
tree = DecisionTreeClassifier(max_depth=50.5)
bagging = BaggingClassifier(base_estimator=a, n_estimators=20, random_state=0, n_jobs=-1)
tree_bag = BaggingClassifier(base_estimator=b, n_estimators=45, random_state=0, n_jobs=-1)
rand_forest = RandomForestClassifier(n_estimators=30, n_jobs=-1)

#т.к. всего 549, на 10 частей - 55
for i in range(1, 11):
    log_reg.fit(X_train_t[:55*i], y_train[:55*i])
    class_names_1 = random_forest.classes_
    y_train_b1 = label_binarize(y_train[:55*i], class_names_1)
    y_test_b1 = label_binarize(y_test, class_names_1)
    y_predict_train_b1 = label_binarize(log_reg.predict(X_train_t[:55*i]), class_names_1)
    y_predict_test_b1 = label_binarize(log_reg.predict(X_test_t), class_names_1)
    
    tree.fit(X_train_t[:55*i], y_train[:55*i])
    class_names_2 = random_forest.classes_
    y_train_b2 = label_binarize(y_train[:55*i], class_names_2)
    y_test_b2 = label_binarize(y_test, class_names_2)
    y_predict_train_b2 = label_binarize(tree.predict(X_train_t[:55*i]), class_names_2)
    y_predict_test_b2 = label_binarize(tree.predict(X_test_t), class_names_2)
    
    bagging.fit(X_train_t[:55*i], y_train[:55*i])
    class_names_3 = random_forest.classes_
    y_train_b3 = label_binarize(y_train[:55*i], class_names_3)
    y_test_b3 = label_binarize(y_test, class_names_3)
    y_predict_train_b3 = label_binarize(bagging.predict(X_train_t[:55*i]), class_names_3)
    y_predict_test_b3 = label_binarize(bagging.predict(X_test_t), class_names_3)
    
    tree_bag.fit(X_train_t[:55*i], y_train[:55*i])
    class_names_4 = random_forest.classes_
    y_train_b4 = label_binarize(y_train[:55*i], class_names_4)
    y_test_b4 = label_binarize(y_test, class_names_4)
    y_predict_train_b4 = label_binarize(tree_bag.predict(X_train_t[:55*i]), class_names_4)
    y_predict_test_b4 = label_binarize(tree_bag.predict(X_test_t), class_names_4)
    
    rand_forest.fit(X_train_t[:55*i], y_train[:55*i])
    class_names_5 = random_forest.classes_
    y_train_b5 = label_binarize(y_train[:55*i], class_names_5)
    y_test_b5 = label_binarize(y_test, class_names_5)
    y_predict_train_b5 = label_binarize(rand_forest.predict(X_train_t[:55*i]), class_names_5)
    y_predict_test_b5 = label_binarize(rand_forest.predict(X_test_t), class_names_5)
    
    acc_train.append(accuracy_score(y_train_b1[:55*i], y_predict_train_b1[:55*i]))
    acc_train.append(accuracy_score(y_train_b2[:55*i], y_predict_train_b2[:55*i]))
    acc_train.append(accuracy_score(y_train_b3[:55*i], y_predict_train_b3[:55*i]))
    acc_train.append(accuracy_score(y_train_b4[:55*i], y_predict_train_b4[:55*i]))
    acc_train.append(accuracy_score(y_train_b5[:55*i], y_predict_train_b5[:55*i]))
    
    f1_train.append(f1_score(y_train_b1[:55*i], y_predict_train_b1[:55*i], average='weighted'))
    f1_train.append(f1_score(y_train_b2[:55*i], y_predict_train_b2[:55*i], average='weighted'))
    f1_train.append(f1_score(y_train_b3[:55*i], y_predict_train_b3[:55*i], average='weighted'))
    f1_train.append(f1_score(y_train_b4[:55*i], y_predict_train_b4[:55*i], average='weighted'))
    f1_train.append(f1_score(y_train_b5[:55*i], y_predict_train_b5[:55*i], average='weighted'))
    
    acc_test.append(accuracy_score(y_test_b1, y_predict_test_b1))
    acc_test.append(accuracy_score(y_test_b2, y_predict_test_b2))
    acc_test.append(accuracy_score(y_test_b3, y_predict_test_b3))
    acc_test.append(accuracy_score(y_test_b4, y_predict_test_b4))
    acc_test.append(accuracy_score(y_test_b5, y_predict_test_b5))
    
    f1_test.append(f1_score(y_test_b1, y_predict_test_b1, average='weighted'))
    f1_test.append(f1_score(y_test_b2, y_predict_test_b2, average='weighted'))
    f1_test.append(f1_score(y_test_b3, y_predict_test_b3, average='weighted'))
    f1_test.append(f1_score(y_test_b4, y_predict_test_b4, average='weighted'))
    f1_test.append(f1_score(y_test_b5, y_predict_test_b5, average='weighted'))
```
